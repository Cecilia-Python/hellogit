```python
import talib as ta
import pandas as pd
import talib.abstract
import warnings
warnings.filterwarnings('ignore')
data = pd.read_excel('sz50.xlsx', sheet_name='600036.XSHG', index_col='datetime')
s = pd.Series(ta.MA(data.close.values, 10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data_dict = {}
data =  pd.read_excel('sz50.xlsx',sheetname='600036.XSHG', index_col='datetime')
data_dict = data['close']
```

```python
rolling_mean = data['close'].rolling(window=10).mean()

plt.figure(figsize=(15, 7))
data['close'].plot()
rolling_mean.plot()
plt.title("10days Rolling Mean of 600036.XSHG")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
